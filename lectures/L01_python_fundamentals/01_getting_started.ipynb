### Order of operations

It's important that we think carefully about the order of operations when we write code as Python will follow the order of operations faithfully...

When we create a new variable through variable assignment, Python does the following:

1. Compute what is to the right of the `=`
2. Store that output in the variable to the left of the `=`

Let's consider an example below where that order of operations might matter:

````python
z = 2

z = z + 4
````

After running this code, what do you think `z` is?

## Code comments

Comments are short notes that you leave for yourself and for others who read your
code.

They should be used to explain what the code does.

A comment is made with the `#`. Python ignores everything in a line that follows a `#`.

Let’s practice making some comments.

````python
i = 1  # Assign the value 1 to variable i
j = 2  # Assign the value 2 to variable j

# We add i and j below this line
i + j
````

## Calling functions

Functions are processes that take an input (or inputs) and produce an output.

If we had a function called `f` that took two arguments `x` and
`y`, we would write `f(x, y)` to use the function.

For example, in our "Hello, world" program, we called the function `print` and passed the variable `x` as an input to that function.

````python
print(x)
````
